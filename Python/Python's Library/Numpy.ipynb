```python
# # Indexing and Slicing
# 1. Slicing
arr_2D = np.random.randint(1, 20, (6, 6))
print(arr_2D)
print()

arr_2D_Sli_1 = arr_2D[:, 0]
print(arr_2D_Sli_1)
print()

arr_2D_Sli_2 = arr_2D[0, :]
print(arr_2D_Sli_2)
print()

arr_2D_Sli_3 = arr_2D[:]
print(arr_2D_Sli_3)
print()

arr_2D_Sli_4 = arr_2D[:, :]
print(arr_2D_Sli_4)
print()

arr_2D_Sli_5 = arr_2D[1:, :2]
print(arr_2D_Sli_5)
print()

arr_2D_Sli_6 = arr_2D[:2, 3:]
print(arr_2D_Sli_6)
print()

arr_2D_Sli_7 = arr_2D[3:][:2]
print(arr_2D_Sli_7)
print()

# 2. Indexing
arr_2D_Ind_1 = arr_2D[0]
print(arr_2D_Ind_1)
print()

arr_2D_Ind_2 = arr_2D[2][0]
print(arr_2D_Ind_2)
print()

arr_2D_Ind_Sli_1 = arr_2D[2][2:]
print(arr_2D_Ind_Sli_1)
print()

arr_2D_Ind_Sli_2 = arr_2D[2:][2]
print(arr_2D_Ind_Sli_2)
print()
```

```python
# # # Perform Basic Operation
# # Performing Some Arithmentic Operation listed below
# # Below opearation perform in 1D Array that is similar to all Dimension.....
arr_1D = np.arange(1, 10)
ar_1D = np.arange(1, 10)
print(arr_1D)
print()
# 1. Addition
arr_1D_add = arr_1D + ar_1D # By using two Array
print(arr_1D_add)
print()

arr_1D_add = arr_1D + [2] # By using Broadcasting
print(arr_1D_add)
print()

arr_1D_add = np.add(arr_1D, ar_1D) # By using two Array
print(arr_1D_add)
print()

arr_1D_add = np.add(arr_1D, [3]) # By using Broadcasting
print(arr_1D_add)
print()

# 2. Subtraction
arr_1D_sub = arr_1D - ar_1D # By using two Array
print(arr_1D_sub)
print()

arr_1D_sub = arr_1D - [2] # By using Broadcasting
print(arr_1D_sub)
print()

arr_1D_sub = np.subtract(arr_1D, ar_1D) # By using two Array
print(arr_1D_sub)
print()

arr_1D_sub = np.subtract(arr_1D, [3]) # By using Broadcasting
print(arr_1D_sub)
print()

# 3. Multiplication
arr_1D_mul = arr_1D * ar_1D # By using two Array
print(arr_1D_mul)
print()

arr_1D_mul = arr_1D * [2] # By using Broadcasting
print(arr_1D_mul)
print()

arr_1D_mul = np.multiply(arr_1D, ar_1D) # By using two Array
print(arr_1D_mul)
print()

arr_1D_mul = np.multiply(arr_1D, [3]) # By using Broadcasting
print(arr_1D_mul)
print()

# 4. Division
arr_1D_div = arr_1D / ar_1D # By using two Array
print(arr_1D_div)
print()

arr_1D_div = arr_1D / [2] # By using Broadcasting
print(arr_1D_div)
print()

arr_1D_div = np.divide(arr_1D, ar_1D) # By using two Array
print(arr_1D_div)
print()

arr_1D_div = np.divide(arr_1D, [3]) # By using Broadcasting
print(arr_1D_div)
print()

# 5. Modulus
arr_1D_mod = arr_1D % ar_1D # By using two Array
print(arr_1D_mod)
print()

arr_1D_mod = arr_1D % [2] # By using Broadcasting
print(arr_1D_mod)
print()

arr_1D_mod = np.mod(arr_1D, ar_1D) # By using two Array
print(arr_1D_mod)
print()

arr_1D_mod = np.mod(arr_1D, [3]) # By using Broadcasting
print(arr_1D_mod)
print()
```

```python
# # Now We Perform some advanced operation
arr_1D = np.arange(1, 10)
ar_1D = np.array([0, np.pi/6, np.pi/4, np.pi/3, np.pi/2])
print(arr_1D)
print()
# 1. cos and cosh
arr_1D_cos = np.cos(ar_1D)
print(arr_1D_cos)
print()

arr_1D_cosh = np.cosh(ar_1D)
print(arr_1D_cosh)
print()

# 2. sin and sinh
arr_1D_sin = np.sin(ar_1D)
print(arr_1D_sin)
print()

arr_1D_sinh = np.sinh(ar_1D)
print(arr_1D_sinh)
print()

# 3. tan and tanh
arr_1D_tan = np.tan(ar_1D)
print(arr_1D_tan)
print()

arr_1D_tanh = np.tanh(ar_1D)
print(arr_1D_tanh)
print()

# 4. exp
arr_1D_exp = np.exp(arr_1D) # e to the power {}
print(arr_1D_exp)
print()

arr_1D_exp2 = np.exp2(arr_1D) # 2 to the power {}
print(arr_1D_exp2)
print()

# 5. log and log10 and log1p and log2
arr_1D_log = np.log(arr_1D)
print(arr_1D_log)
print()

arr_1D_log10 = np.log10([10, 100, 1000, 10000, 100000])
print(arr_1D_log10)
print()

arr_1D_log1p = np.log1p(arr_1D) # perform log(exp() - 1)
print(arr_1D_log1p)
print()

arr_1D_log2 = np.log2([2, 4, 8, 16, 32])
print(arr_1D_log2)
print() 
```

```python
# # # I/O Operation
# # Using Delimiter
from io import StringIO
data = u"1, 2, 3\n4, 5, 6"
arr_IO_1 = np.genfromtxt(StringIO(data), delimiter=",")
print(arr_IO_1)
print()

data = u"  1  2  3\n  4  5 67\n890123  4"
arr_IO_2 = np.genfromtxt(StringIO(data), delimiter=3)
print(arr_IO_2)
print()

data = u"123456789\n   4  7 9\n   4567 9"
arr_IO_3 = np.genfromtxt(StringIO(data), delimiter=(4, 3, 2))
print(arr_IO_3)
print()

# # Using autostrip
data = u"1, abc , 2\n 3, xxx, 4"
# Without autostrip
arr_IO_4 = np.genfromtxt(StringIO(data), delimiter=",", dtype="|U5")
print(arr_IO_4)
print()

# With autostrip
arr_IO_5 = np.genfromtxt(StringIO(data), delimiter=",", dtype="|U5", autostrip=True)
print(arr_IO_5)
print()

# # Using comments
data = u"""#
# Skip me !
# Skip me too !
1, 2
3, 4
5, 6 #This is the third line of the data
7, 8
# And here comes the last line
9, 0
"""
arr_IO_6 = np.genfromtxt(StringIO(data), comments="#", delimiter=",")
print(arr_IO_6)
print()

# # skip_header and skip_footer
data = u"\n".join(str(i) for i in range(10))
arr_IO_7 = np.genfromtxt(StringIO(data),)
print(arr_IO_7)
print()

arr_IO_7 = np.genfromtxt(StringIO(data),skip_header=3, skip_footer=5)
print(arr_IO_7)
print()

# # Useing usecols
data = u"1 2 3\n4 5 6"
arr_IO_8 = np.genfromtxt(StringIO(data), usecols=(0, -1))
print(arr_IO_8)
print()

arr_IO_9 = np.genfromtxt(StringIO(data),names="a, b, c", usecols=("a", "c"))
print(arr_IO_9)
print()

arr_IO_10 = np.genfromtxt(StringIO(data),names="a, b, c", usecols=("a, c"))
print(arr_IO_10)
print()
```

```python
# # Use Names
data = StringIO("1 2 3\n 4 5 6")
arr_IO_11 = np.genfromtxt(data, names="A, B, C")
print(arr_IO_11)
print()

data = StringIO("So it goes\n#a b c\n1 2 3\n 4 5 6")
arr_IO_12 = np.genfromtxt(data, skip_header=1, names=True)
print(arr_IO_12)
print()


data = StringIO("1 2 3\n 4 5 6")
ndtype = [('a',int), ('b', float), ('c', int)]
names = ["A", "B", "C"]
arr_IO_13 = np.genfromtxt(data, names=names, dtype=ndtype)
print(arr_IO_13)
print()

# # Lets practice with converters arguments
convertfunc = lambda x: float(x.strip(b"%"))/100.
data = u"1, 2.3%, 45.\n6, 78.9%, 0"
names = ("i", "p", "n")
# General case
arr_IO_14 = np.genfromtxt(StringIO(data), delimiter=",", names=names)
print(arr_IO_14)
print()

# Converted case
arr_IO_15 = np.genfromtxt(StringIO(data), delimiter=",", names=names,converters={1: convertfunc})
print(arr_IO_15)
print()
```
